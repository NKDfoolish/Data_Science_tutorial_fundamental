## Pyplot API vs Object Oriented (OO) API

+ Pyplot API > quickly
+ OO API > Advanced

### Pyplot API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = 'blue', linestyle = 'dashed', label = 'sin(x)');
plt.plot(x, np.cos(x), color = 'red', label = "cos(x)")

plt.title('Sin(X) and Cos(X)')
plt.xlabel("X");
plt.ylabel("F(X)");

#plt.xlim([0,4])
#plt.ylim([-0.6,0.6])
#plt.axis([0,4,-0.6,0.6])

#plt.axis('tight')
plt.axis('equal')

plt.legend();
```

### Object - Oriented API

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # Figure size = Width and Height of the plot

#3. Plot the data
ax.plot(x,y);

ax.set(title='A simple plot',
      xlabel='x-axis',
      ylabel='y-axis');
```

## 2. Most common types of Matplotlib plots
+ line
+ scatter
+ bar
+ hist
+ subplots()

### line

```python
# create an array
x = np.linspace(0,10,100)
x[:5]
```

```python
##OO API plot a line chart
fig, ax = plt.subplots()
ax.plot(x,x**3);
```

### scatter

```python
# Pyplot API
plt.scatter(x,np.exp(x)); # y = e^(x)
```

```python
# OO API
# Prepare data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s=sizes, c=colors, cmap='viridis', alpha=0.3);

fig.colorbar(img1);
```

### 3.Bar
+ vertical
+ horizontal

```python
# prepare data from a dictionary
soft_drink_price = {"Coke":10,
                   "Pessi":12,
                   "Sprite":8}
fig, ax = plt.subplots()
ax.bar(soft_drink_price.keys(), soft_drink_price.values());
ax.set(title="Bach Hoa Xanh's Soft Drink Price", ylabel="Price ($)" )
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_price.keys()), list(soft_drink_price.values()));
```

### 4.Histogram

```python
# prepare data
np.random.seed(42)
student_height = np.random.normal(170,10,250)
```

```python
student_height[:10], len(student_height)
```

```python
# pyplot API
plt.hist(student_height, bins=30);
```

### Subplots

```python
# Option 1:
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

# Plot data to each axis
ax1.plot(x,x/2);
ax1.plot(x,x*2);

ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(soft_drink_price.keys(), soft_drink_price.values());
ax4.hist(student_height, bins=30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

# Plot data to each axis
ax[0,0].plot(x,x/2);
ax[0,0].plot(x,x*2);

ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(soft_drink_price.keys(), soft_drink_price.values());
ax[1,1].hist(student_height, bins=30);
```

### Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("./data/california_cities.csv")
cities.head()
```

```python
# extract latd = vĩ độ; longd = kinh độ
lat, lon = cities['latd'], cities['longd']

plt.style.use('seaborn')
plt.figure(figsize=(9,6))

population, area = cities['population_total'], cities['area_total_km2']
# plot using pyplot API
plt.scatter(lon,lat, c=np.log10(population), cmap='viridis', s=area)

plt.axis("equal")
plt.xlabel('longlatitude')
plt.ylabel('latitude');

plt.colorbar(label='log$_{10}$(population)');
plt.clim(3,7);


# create a legend cities's size
area_range = [50,100,300,500]

for area in area_range:
    plt.scatter([],[], 
                s=area, 
                label=str(area)+'km$^2$', c='k', alpha=0.4)
plt.legend(labelspacing=1, title='City Area') # labelspacing : khoảng cách giữa các label

plt.title('California Cities: Population and Distribution');
```
